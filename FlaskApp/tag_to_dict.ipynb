```python
%%time
import lxml.etree as ET

tags = {}

def tag_iter(context, *args, **kwargs):
    tags = {}
    for event, elem in context:
        try:
            tags[elem.attrib['Id']] = elem.attrib['TagName']
        except Exception as e:
            print(elem.attrib)
        elem.clear()
        for ancestor in elem.xpath('ancestor-or-self::*'):
            while ancestor.getprevious() is not None:
                del ancestor.getparent()[0]
    del context
    return tags

context = ET.iterparse('data/Tags.xml', events=('end', ))
tags = tag_iter(context)
```
